```python
import pandas as pd 
import requests as req
from bs4 import BeautifulSoup 
import re 
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
final = pd.DataFrame()

for j in range(151,300):
    
    url  = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Referer": "https://www.google.com/",
    "Connection": "keep-alive"
}
    webpage = req.get(url, headers = headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    company = soup.find_all(class_='companyCardWrapper')

    # adding the parameter in the data frame 

    name = []
    company_rating = []
    companyRatingCount = []
    links = []
    founded_year = []
    headquater = []
    employees = []

    # --- Step 2: Extract basic info + links ---
    for i in company:
        name.append(i.find('h2').text.strip() if i.find('h2') else None)
        
        rating_tag = i.find('div', class_='rating_text')
        company_rating.append(rating_tag.text.strip() if rating_tag else None)
        
        count_tag = i.find('span', class_="companyCardWrapper__companyRatingCount")
        companyRatingCount.append(count_tag.text.strip().replace("(", "").replace(")", "") if count_tag else None)
        
        link_tag = i.find('meta', itemprop='url')
        links.append(link_tag.get('content') if link_tag else None)

    # Helper Functions to convert the string to number 

    def convert_to_number(text):
        match = re.search(r'(\d+)', text)
        if not match:
            return None
        
        num = int(match.group())
        
        if 'lakh' in text:
            return num * 100000
        elif 'k' in text:
            return num * 1000
        else:
            return num
    
    def clean_employee(text):
        if not text:
            return None
        
        text = text.lower().replace(',', '').strip()
        
        # Handle range like "50k-1 lakh"
        if '-' in text:
            parts = text.split('-')
            values = [convert_to_number(p) for p in parts]
            return max(values) if all(values) else None
        
        return convert_to_number(text)


        # --- Step 3: Visit each company page ---
    for link in links:
        if link:
            page = req.get(link, headers=headers).text
            soup2 = BeautifulSoup(page, 'lxml')
            
            # -------- Founded Year --------
            founded_val = None
            label = soup2.find('div', string=lambda x: x and 'Founded' in x)
            
            if label:
                next_div = label.find_next_sibling('div')
                prev_div = label.find_previous_sibling('div')
                
                text = (next_div.text if next_div else '') or (prev_div.text if prev_div else '')
                
                match = re.search(r'\d{4}', text)
                founded_val = match.group() if match else None
            
            founded_year.append(founded_val)
            
            # -------- Headquarters --------
            hq_label = soup2.find('div', string=lambda x: x and 'Headquarters' in x)        
            if hq_label:
                # Try both cases: direct div OR inside <a>
                hq_value = hq_label.find_next_sibling('div') or hq_label.find_next('a')           
                headquater.append(hq_value.text.strip() if hq_value else None)
            else:
                headquater.append(None)
    
            # -------- Employees --------
            emp_val = None
            
            emp_label = soup2.find('div', string=lambda x: x and ('Employee Count' in x or 'Company Size' in x ))
            
            if emp_label:
                next_div = emp_label.find_next_sibling('div')
                prev_div = emp_label.find_previous_sibling('div')
                
                text = (next_div.text if next_div else '') or (prev_div.text if prev_div else '')
                
                emp_val = clean_employee(text)
            
            employees.append(emp_val)
            
        else:
            founded_year.append(None)
            headquater.append(None)
            employees.append(None)

    # addign to the data frame 
    
    df = pd.DataFrame({
    "Name": name,
    "Rating": company_rating,
    "Rating Count": companyRatingCount,
    "Headquarters": headquater,
    "Company_year" : founded_year,
    "Employees Count"  : employees
    })

    # concat each df in final data frame 
    
    final = pd.concat([final, df], ignore_index=True)
```
